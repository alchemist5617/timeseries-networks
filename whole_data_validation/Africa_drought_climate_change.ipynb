```python
def drougth_counter(d, level = -1):
    level = -1
    N = 0
    in_drought = False
    for i in range(1,len(d)):
        if d[i-1] > level and d[i] <= level and not in_drought:
            N = N+1
            in_drought = True
        if in_drought and d[i-1] < 0 and d[i] >= 0:
            in_drought = False
    return(N)
```

```python
def data_cleaner_zero(data):
    for k in range(data.shape[1]):
        a = data[:,k]
        a[np.isnan(a)] = 0
```

```python
def data_cleaner_repl(data):
    data_clean = np.zeros(data.shape)
    for k in range(data.shape[1]):
        a = data[:,k]
        for i in range(len(a)):
            if np.isnan(a[i]):
                if i==0:
                    start = 0
                else:
                    start = a[i-1]        
                j = 0
                while np.isnan(a[i+j+1]):
                    j += 1
                    if i+j+1 == len(a):
                        j -= 1
                        break
                end = a[i+j+1]
                inc = (end - start)/(j+2)
                for k in range(j):
                    a[i+k] = start + (k+1)*inc
                
        data_clean[:,k] = a
```

```python
def data_cleaner(data):
    data_clean = np.zeros(data.shape)
    for k in range(data.shape[1]):
        a = data[:,k]
        for i in range(len(a)):
            if np.isnan(a[i]):
                if i==0:
                    start = 0
                else:
                    start = a[i-1]        
                j = 0
                while np.isnan(a[i+j+1]):
                    j += 1
                    if i+j+1 == len(a):
                        j = j - 1
                        break;
                end = a[i+j+1]
                inc = (end - start)/(j+2)
                for k in range(j):
                    a[i+k] = start + (k+1)*inc             
        data_clean[:,k] = a
    return(data_clean)
```

```python
def neighbour_vote_class(dic,lat,lon):
    shift = [[-0.25,0.25],[-0.25,0.25]]
    shift_list = []
    for element in itertools.product(*shift):
        shift_list.append(element)
    #shift_list.remove((0,0))
    result = []
    for x,y in shift_list:
        if lon > 180: lon -= 360
        if (lat + y, lon + x) in class_dic.keys():
            result.append(class_dic[(lat + y, lon + x)])   
    return(np.array(result))
```
